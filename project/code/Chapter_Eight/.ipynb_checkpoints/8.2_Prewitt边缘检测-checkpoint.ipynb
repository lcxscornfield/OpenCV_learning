Roberts算法本质是对计算对角线上的差分值来获得边缘强度。且容易受到噪点干扰。为解决这个问题，提出Prewitt算法。

Prewitt算法将可分离卷积核分离为：均值平滑算子 ★ 差分算子的形式。既解决了噪点问题，也解决了边缘检测问题。
* prewittx反应垂直方向边缘
* prewitty反应水平方向边缘
* 边缘强度衡量方式选择插值法或平方和开方法。

Prewitt算子的实现

```python
import numpy as np
from scipy import signal
def prewitt(I,_boundary='symm',):
    #prewitt_x可分离，可以分两次卷积操作
    #1.垂直方向均值平滑
    ones_y = np.array([[1],[1],[1]],np.float32)
    i_conv_pre_x = signal.convolve2d(I,ones_y,mode='same',boundary = _boundary)
    #2.水平方向差分
    deff_x = np.array([[1,0,-1]],np.float32)
    i_conv_pre_x = signal.convolve2d(i_conv_pre_x,deff_x,mode='same',boundary = _boundary)
    
    #prewitty_y进行同样步骤
    ones_x = np.array([[1,1,1]],np.float32)
    i_conv_pre_y = signal.convolve2d(I,ones_x,mode='same',boundary = _boundary)
    deff_y = np.array([[1],[0],[-1]],np.float32)
    i_conv_pre_y = signal.convolve2d(i_conv_pre_y,deff_y,mode='same',boundary = _boundary)
    return (i_conv_pre_x,i_conv_pre_y)
```

调用Prewitt算子

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

src = cv2.imread("img7.jpg",0)
#求prewittx&prewitty算子的边缘强度和灰度级
i_conv_pre_x,i_conv_pre_y = prewitt(src)
#取绝对值求边缘强度
abs_i_conv_pre_x = np.abs(i_conv_pre_x)
abs_i_conv_pre_y = np.abs(i_conv_pre_y)
#求灰度级
edge_x = abs_i_conv_pre_x.copy()#用拷贝，因为abs之后还要用于插值法
edge_y = abs_i_conv_pre_y.copy()
edge_x[edge_x>255]=255
edge_y[edge_y>255]=255
#数据类型转化
edge_x = edge_x.astype(np.uint8)
edge_y = edge_y.astype(np.uint8)
#cv2.imshow("edge_x",edge_x)
#cv2.imshow("edge_y",edge_y)

#插值法interpolation
edge_itp = 0.5*abs_i_conv_pre_x + 0.5*abs_i_conv_pre_y
edge_itp[edge_itp>255] = 255
edge_itp = edge_itp.astype(np.uint8)

#平方开方法
edge_sqrt = np.sqrt(np.power(abs_i_conv_pre_x,2.0)+np.power(abs_i_conv_pre_y,2.0))
edge_sqrt = np.round(edge_sqrt) #np.round为小数点四舍五入，默认保留0位
edge_sqrt[edge_sqrt>255] = 255
edge_sqrt = edge_sqrt.astype(np.uint8)

titles = ["src","edge_x","edge_y","edge_itp","edge_sqrt"]
images = [src,edge_x,edge_y,edge_itp,edge_sqrt]
for i in xrange(5):
    plt.subplot(3,2,i+1)
    cv2.imshow(titles[i],images[i])
    plt.title(titles[i])
    plt.imshow(images[i])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

edge_x,edge_y很好的检测到垂直方向与竖直方向的边缘
